## Project: Basic Data Visualization


In this task, I worked with the Iris dataset, which includes measurements of different flower species.  
My goal was to:
- Create basic plots (bar chart, line chart, and scatter plot)
- Customize the visuals with proper titles, labels, and legends
- Save each plot as an image for reporting purposes

The tools I used include Python, Matplotlib, and Seaborn

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# Loading the Iris dataset
df = pd.read_csv('iris.csv')
```

```python
# Previewing the data
print(f'Display the first few rows of the dataset')
df.head()
```

```python
# A Bar plot showing average petal length for each Iris species
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='petal_length', data=df)
plt.title("Average Petal Length per Iris Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.savefig("barplot_petal_length.png")  # Saves the plot as an image
plt.show()
```

```python
# Line plot showing sepal length of the first 30 flowers
plt.figure(figsize=(10, 5))
plt.plot(df['sepal_length'][:30], marker='o', linestyle='-', color='green')
plt.title("Sepal Length of First 30 Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.savefig("lineplot_sepal_length.png")
plt.show()
```

```python
# Scatter plot comparing petal length and width, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title='Species')
plt.savefig("scatterplot_petal.png")
plt.show()
```
